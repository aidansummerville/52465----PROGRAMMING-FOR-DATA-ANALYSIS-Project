```python
ndf = pd.DataFrame(breed, columns = ['Breed'])

No_Owners = []


# For each row in the column,
for row in ndf['Breed']:
    # if more than a value,
    if row == "AAX":
        a = np.random.choice(no, p=[0.714, 0.286])
        No_Owners.append(a)
    # else, if more than a value,
    elif row == "LMX":
        b = np.random.choice(no2,p=[41/49, 6/49, 2/49])
        # Append a letter grade
        No_Owners.append(b)
    # else, if more than a value,
    elif row == "CHX":
        c = np.random.choice(no,p=[24/26, 2/26])
        # Append a letter grade
        No_Owners.append(c)

# Create a column from the list
ndf['No.Owners'] = No_Owners

ndf
```
